````python
import tensorflow as tf
import matplotlib.pyplot as plt
````

````python
IMAGE_SIZE = (299, 299)
DATASET_PATH = '../input/weather-dataset/dataset'
SEED = 100
````

https://www.tensorflow.org/api_docs/python/tf/keras/utils/image_dataset_from_directory?version=nightly

````python
train_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_PATH,
    batch_size=32,
    image_size=IMAGE_SIZE,
    shuffle=True,
    seed=SEED,
    validation_split=0.3,
    subset='training',
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    DATASET_PATH,
    batch_size=32,
    image_size=IMAGE_SIZE,
    shuffle=True,
    seed=SEED,
    validation_split=0.3,
    subset='validation',
)
````

````python
class_names = train_ds.class_names
class_names
````

````python
def preprocess(image, label):
    image = tf.keras.applications.inception_v3.preprocess_input(image)
    return image, label

train_ds = train_ds.map(preprocess)
val_ds = val_ds.map(preprocess)
````

````python
feature_extractor = tf.keras.applications.inception_v3.InceptionV3(
    include_top=False,
    weights='imagenet',
)
````

````python
x = tf.keras.layers.GlobalAveragePooling2D()(feature_extractor.outputs[0])
x = tf.keras.layers.Flatten()(x)
output = tf.keras.layers.Dense(len(class_names), activation='softmax')(x)

model = tf.keras.Model(inputs = feature_extractor.inputs, outputs = [output])
model.summary()
````

````python
feature_extractor.trainable = False

model.compile(
    loss='sparse_categorical_crossentropy',\
    optimizer='adam',
    metrics = ['accuracy']
)
````

````python
model.fit(train_ds, validation_data=val_ds, epochs=2)
````
