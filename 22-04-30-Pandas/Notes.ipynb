# Series = 1D

````python
s = pd.Series(['Diwakar Gupta', 19, 5.5], index=['Name', 'Age', 'Height'])
````

````python
print(s)
````

````python
print(s['Name'])
print(s['Age'])
print(s['Height'])
````

````python
s['Name'] = 'Diwakar'
print(s)
````

````python
print(s.values)
print(type(s.values))
print(s.index)
print(type(s.index))
````

````python
A = pd.Series([23, 19, 5.5]) # implicit indexing
A
````

````python
A.min(), A.values.min(), A.max(), A.mean(), A.size, A.shape
````

## Indexing


*   Implicit: 0, 1, 2, 3  [included, excluded]
*   Explicit: `index=['Name', 'Age']`  [included, included]



````python
A[1]
````

````python
print(A)
print()
print(A[0:2])
````

````python
print(s) # Explicit indexing
print()
print(s['Name':'Height']) # s[include : include]
print()
print(s[0:2]) # implicit: s[included : excluded]
````

````python
print(s)
print()
print(s.iloc[0:2]) # implicit 0-based indexing
print()
print(s.loc['Age':'Height']) # explicit: user-defined indexing
````

````python
print(A)
print()
A[A <= 20]
````
