```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

# Relplot

https://seaborn.pydata.org/api.html



*   scatter
*   line
*   



```python
# hue -> color

# sns.relplot(x=[1, 3, 6, 2], y=[7,8,4,5])

# sns.relplot(x=tips['total_bill'], y=tips['tip'])
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size', kind='line')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', col='time')
```

# Catplot



*   swarm
*   strip
*   count
*   box
*   violin



```python
tips.sample(3)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', jitter=True)
```

```python
# sns.set_theme(style="ticks")

sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='swarm')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='strip')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='boxen')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
```

```python
titanic = sns.load_dataset('titanic')
titanic.sample(5)
```

```python
sns.catplot(data=titanic, x='alive', kind='count', hue='sex')
```

```python
sns.catplot(data=titanic, x='alive', kind='count', hue='pclass')
```

palette='Set3' # color scheme

```python
sns.set_theme(style="darkgrid")

sns.catplot(data=titanic, x='embarked', kind='count', hue='pclass')
```

# Distribution



*   kde
*   jointplot
    *   hex
    *   kde
*   pairplot



```python
sns.displot(data=titanic, x='age', kde=True, hue='alive')
```

```python
sns.kdeplot(data=titanic, x='age', hue='alive')
```

```python
sns.displot(data=titanic, x='age', kde=True, hue='alive', col='sex')
```

# JointPlot

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```
